```python
def get_coauthorlist(AuthorID,year_paper_list,n):
    coauthorlist = []
    count = 0
    for pair in year_paper_list: 
        count+=1
        coauthorlist.extend(paperAuthorDict[pair[1]].split(','))
        if count == n:
            break
    coauthorlist = set(coauthorlist)
    coauthorlist.remove(AuthorID)
    coauthorlist = list(coauthorlist)
    return coauthorlist

def get_seqlist(a):
    return authorSeqDict[a].split(',')

def get_yearlist(paperlist):
    yearlist=[]
    for p in paperlist:
        yearlist.append(int(paperYearDict[p]))
    yearlist = sorted(yearlist)
    return yearlist

def get_student_list(MID):
    return mentorStudentDict[MID]

def get_student_num(MID):
    return len(mentorStudentDict[MID])

def get_student_achievements(MID):
    ach = 0.0
    for s in mentorStudentDict[MID]:
        s_seqlist = get_seqlist(s)
        for p in s_seqlist:
            ach += 1.0/int(p)
    return ach
```
